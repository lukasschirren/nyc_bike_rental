```python
##SATURDAYS

# Convertir la colonne de date au format datetime
data_2019['starttime'] = pd.to_datetime(data_2019['starttime'])

# Extraire le jour de la semaine
data_2019['day_of_week'] = data_2019['starttime'].dt.dayofweek  # 0: Lundi, 1: Mardi, ..., 5: Samedi, 6: Dimanche

# Filtrer les donnÃ©es pour inclure uniquement les samedis (jour_de_la_semaine = 5)
saturdays_data = data_2019[data_2019['day_of_week'] == 5]

# Compter le nombre d'occurrences de chaque station en tant que "start station name" sur les samedis
saturdays_station_counts = saturdays_data['start station name'].value_counts()

# Calculer le nombre de samedis de l'annÃ©e 2019
num_saturdays_2019 = len(data_2019[data_2019['day_of_week'] == 5]['starttime'].dt.date.unique())

# Calculer le nombre moyen de dÃ©parts par station le samedi pour chaque station
average_saturdays_per_station = (saturdays_station_counts/num_saturdays_2019).round(0).astype(int)

# Afficher le rÃ©sultat
print("10 first stations with the highest demand (in terms of starting stations) on Saturdays")
print(average_saturdays_per_station.head(10))

```

```python
##SUNDAYS

# Convertir la colonne de date au format datetime
data_2019['starttime'] = pd.to_datetime(data_2019['starttime'])

# Extraire le jour de la semaine
data_2019['day_of_week'] = data_2019['starttime'].dt.dayofweek  # 0: Lundi, 1: Mardi, ..., 5: Samedi, 6: Dimanche

# Filtrer les donnÃ©es pour inclure uniquement les dimanche
sundays_data = data_2019[data_2019['day_of_week'] == 6]

# Compter le nombre d'occurrences de chaque station en tant que "start station name" sur les samedis
sundays_station_counts = sundays_data['start station name'].value_counts()

# Calculer le nombre de samedis de l'annÃ©e 2019
num_sundays_2019 = len(data_2019[data_2019['day_of_week'] == 6]['starttime'].dt.date.unique())

# Calculer le nombre moyen de dÃ©parts par station le samedi pour chaque station
average_sundays_per_station = (sundays_station_counts/num_sundays_2019).round(0).astype(int)

# Afficher le rÃ©sultat
print("10 first stations with the highest demand (in terms of starting stations) on Saturdays")
print(average_sundays_per_station.head(10))
```

```python
#WEEKDAYS
# Convertir la colonne de date au format datetime
data_2019['starttime'] = pd.to_datetime(data_2019['starttime'])

# Extraire le jour de la semaine
data_2019['day_of_week'] = data_2019['starttime'].dt.dayofweek  # 0: Lundi, 1: Mardi, ..., 6: Dimanche

# Filtrer les donnÃ©es pour inclure uniquement les jours de la semaine (lundi Ã  vendredi)
weekdays_data = data_2019[(data_2019['day_of_week'] >= 0) & (data_2019['day_of_week'] <= 4)]

# Compter le nombre d'occurrences de chaque station en tant que "start station name" sur les jours de la semaine
weekdays_station_counts = weekdays_data['start station name'].value_counts()

# Calculer le nombre de jours de la semaine (lundi Ã  vendredi) dans l'annÃ©e 2019
num_weekdays_2019 = len(weekdays_data['starttime'].dt.date.unique())

# Calculer le nombre moyen de dÃ©parts par station pour chaque jour de la semaine (lundi Ã  vendredi)
average_weekdays_per_station = weekdays_station_counts / num_weekdays_2019

# Arrondir chaque valeur Ã  l'entier le plus proche
average_weekdays_per_station = average_weekdays_per_station.round(0).astype(int)

# Afficher le rÃ©sultat
print("Top 10 stations with the highest demand per day (in terms of starting stations) during the week")
print(average_weekdays_per_station)
```
