```python
##SATURDAYS - START STATIONS

# Convertir la colonne de date au format datetime
data_2019['starttime'] = pd.to_datetime(data_2019['starttime'])

# Extraire le jour de la semaine
data_2019['day_of_week'] = data_2019['starttime'].dt.dayofweek  # 0: Lundi, 1: Mardi, ..., 5: Samedi, 6: Dimanche

# Filtrer les donnÃ©es pour inclure uniquement les samedis (jour_de_la_semaine = 5)
saturdays_data = data_2019[data_2019['day_of_week'] == 5]

# Compter le nombre d'occurrences de chaque station en tant que "start station name" sur les samedis
saturdays_station_counts = saturdays_data['start station name'].value_counts()

# Calculer le nombre de samedis de l'annÃ©e 2019
num_saturdays_2019 = len(data_2019[data_2019['day_of_week'] == 5]['starttime'].dt.date.unique())

# Calculer le nombre moyen de dÃ©parts par station le samedi pour chaque station
average_saturdays_per_station = (saturdays_station_counts/num_saturdays_2019).round(0).astype(int)

# Afficher le rÃ©sultat
print("10 first stations with the highest demand (in terms of starting stations) on Saturdays")
print(average_saturdays_per_station.head(10))

```

```python
##2019 - WEEKDAYS - BETWEEN 6AM AND 10AM AND BETWEEN 4PM AND 8PM
# Supposons que data_2019 soit votre DataFrame contenant les colonnes "start station name" et "starttime"

# Convertir la colonne de date au format datetime
data_2019['starttime'] = pd.to_datetime(data_2019['starttime'])

# Extraire le jour de la semaine
data_2019['day_of_week'] = data_2019['starttime'].dt.dayofweek  # 0: Lundi, 1: Mardi, ..., 6: Dimanche

# Filtrer les donnÃ©es pour inclure uniquement les jours de la semaine (lundi Ã  vendredi) et les heures entre 7h et 9h
weekday_morning_data = data_2019[(data_2019['day_of_week'] >= 0) & (data_2019['day_of_week'] <= 4) & (data_2019['starttime'].dt.hour >= 6) & (data_2019['starttime'].dt.hour < 10)]

# Compter le nombre d'occurrences de chaque station en tant que "start station name" pour les jours de la semaine et les heures spÃ©cifiÃ©es
weekday_morning_station_counts = weekday_morning_data['start station name'].value_counts()

# Calculer le nombre de jours de la semaine (lundi Ã  vendredi) dans l'annÃ©e 2019
num_weekdays_2019 = len(weekday_morning_data['starttime'].dt.date.unique())

# Calculer le nombre moyen de dÃ©parts par station pour les jours de la semaine et les heures spÃ©cifiÃ©es
average_weekday_morning = weekday_morning_station_counts / num_weekdays_2019

# Arrondir chaque valeur Ã  l'entier le plus proche
average_weekday_morning = average_weekday_morning.round(0).astype(int)

# Afficher le rÃ©sultat
print("Average number of departures per station for days of the week and hours between 6 a.m. and 10 a.m. in 2019:")
print(average_weekday_morning.head(10))


weekday_evening_data = data_2019[(data_2019['day_of_week'] >= 0) & (data_2019['day_of_week'] <= 4) & (data_2019['starttime'].dt.hour >= 16) & (data_2019['starttime'].dt.hour < 20)]

weekday_evening_station_counts = weekday_evening_data['start station name'].value_counts()

average_weekday_evening = weekday_evening_station_counts / num_weekdays_2019

average_weekday_evening = average_weekday_evening.round(0).astype(int)

# Afficher le rÃ©sultat
print("Average number of departures per station for days of the week and hours between  4p.m. and 8 p.m. in 2019:")
print(average_weekday_evening.head(10))
```
